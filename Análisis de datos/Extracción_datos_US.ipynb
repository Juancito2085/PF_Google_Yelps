```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
# URL de la página
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States_by_population_density"

# Hacer una solicitud GET a la página
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
 table = soup.find('table', {'class': 'wikitable'})
```

```python
#itero sobre table
rows = table.find_all('tr')
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) 
    #se borran los dos elementos de data
data.pop(0)
data.pop(0)
#se itera sobre data para sacar la primera y la tercera columna
data2 = []
for i in data:
    data2.append([i[0], i[2]])
#se asigna el primer elemento a la lista densidad y el segundo a la lista poblacion
densidad = []
poblacion = []
for i in data2:
    densidad.append(i[0])
    poblacion.append(i[1])

#ahora se recorre la tabla para sacar los nombres de los estados
rows = table.find_all('tr')
data = []
for row in rows:
    cols = row.find_all('th')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) 
    #se borran los dos elementos de data
data.pop(0)
data.pop(0)
#Se crea una lista con los nombres de los estados
estados = []
for i in data:
    estados.append(i[0])
```

```python
#Se crea un diccionario con los datos
datos = {'Estado':estados, 'Densidad':densidad, 'Poblacion':poblacion}
```

```python
df=pd.DataFrame(datos)
```

```python
df
```

```python
df=df[~df['Estado'].isin(['District of Columbia','Puerto Rico','American Samoa','U.S. Virgin Islands',
                                  'Northern Mariana Islands','Guam','50 States','50 States and DC','United States','Contiguous US'])]
```

```python
df.to_csv('Fuentes externas/estados_densidad_poblacion1.csv', index=False, encoding='utf-8')
```
