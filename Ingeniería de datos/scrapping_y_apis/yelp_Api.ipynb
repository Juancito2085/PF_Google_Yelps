```python
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    'Houston': {'auto repair': 20, 'gas stations': 15, 'restaurants': 50, 'beauty salons': 30, 'churches': 10},
    'New York': {'auto repair': 25, 'gas stations': 20, 'restaurants': 60, 'beauty salons': 40, 'churches': 15},
    # Agrega más datos para otras ciudades...
}

cities = list(data.keys())
categories = list(data[cities[0]].keys())

# Crear gráficos de barras para cada categoría
for category in categories:
    counts = [data[city][category] for city in cities]
    plt.figure(figsize=(10, 5))
    plt.bar(cities, counts, color='blue')
    plt.title(f'Número de {category} en diferentes ciudades')
    plt.xlabel('Ciudad')
    plt.ylabel('Número de Negocios')
    plt.show()
```
